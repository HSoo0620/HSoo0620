```python
from scipy.interpolate import NearestNDInterpolator
import matplotlib.pyplot as plt
import nibabel as nib
from skimage.transform import resize
import itertools
from tensorflow.keras import backend as K
import nibabel as nib
import os 
import numpy as np
```

```python
#기존의 one-hot encoding을 했을 경우 

filename = "./BTCV_Data/Rigid_gt/1.nii.gz"
x = nib.load(filename).get_fdata()

x = x.astype(np.int32)

x = K.one_hot(K.cast(x, 'int32'), num_classes=14)[..., 1:]

x = x.numpy()

savename = "./BTCV_Data/1.test.nii.gz"
x = nib.Nifti1Image(x, None) 
nib.save(x, savename)
```
