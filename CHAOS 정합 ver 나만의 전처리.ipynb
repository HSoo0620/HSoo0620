```python
# windowing [-200,500]
# DCM Slice 들을 Voxel 뭉텅이로 모두 바꾸는 코드

for a in range(0,31):
    voxel = "./chaos_train_voxel/원본사이즈/" + str(a) + ".nii.gz"
    if os.path.isfile(voxel):
        voxel = nib.load(voxel).get_fdata()

        for i in range(voxel.shape[0]):
            for j in range(voxel.shape[1]):
                for k in range(voxel.shape[2]):
                    if (voxel[i][j][k] < -200):
                        voxel[i][j][k] = 0
                    if (voxel[i][j][k] > 500):
                        voxel[i][j][k] = 1

        savename = "./chaos_train_voxel/windowing/" + str(a) + ".nii.gz"
        x = nib.Nifti1Image(voxel, None) 

        nib.save(x, savename)
        print(x.shape)
    
```

```python
# Resize
def resize_data(data):
    initial_size_x = data.shape[0]
    initial_size_y = data.shape[1]
    initial_size_z = data.shape[2]

    new_size_x = 128
    new_size_y = 128
    new_size_z = 64

    delta_x = initial_size_x / new_size_x
    delta_y = initial_size_y / new_size_y
    delta_z = initial_size_z / new_size_z

    new_data = np.zeros((new_size_x, new_size_y, new_size_z))

    for x, y, z in itertools.product(range(new_size_x),
                                     range(new_size_y),
                                     range(new_size_z)):
        new_data[x][y][z] = data[int(x * delta_x)][int(y * delta_y)][int(z * delta_z)]

    return new_data
```

```python
# 2windowing 을 3 resize로 
a = 1

for i in range (0, 31):
    filename = "./chaos_train_voxel/2windowing/" + str(a) + ".nii.gz"
    if os.path.isfile(filename):
        x = nib.load(filename).get_fdata()
        print(a, ":", x.shape )
        x = resize_data(x)
        print(a, ":", x.shape )
        savename = "./chaos_train_voxel/3resize/" + str(a) + ".nii.gz"
        
        x = nib.Nifti1Image(x, None) 
        nib.save(x, savename)
#         print(a, ":", x.shape )
    
    a+=1
```

```python
# 2windowing 을 3 resize로 
a = 1

for i in range (0, 31):
    filename = "./chaos_train_voxel/gt/" + str(a) + ".nii.gz"
    if os.path.isfile(filename):
        x = nib.load(filename).get_fdata()
        print(a, ":", x.shape )
        x = resize_data(x)
        print(a, ":", x.shape )
        savename = "./chaos_train_voxel/gt_resize/" + str(a) + ".nii.gz"
        
        x = nib.Nifti1Image(x, None) 
        nib.save(x, savename)
#         print(a, ":", x.shape )
    
    a+=1
```
